```python
import torch
import os
import gr00t
import datetime

PATH = os.path.join(os.path.dirname(os.path.dirname(gr00t.__file__)),"collect_data/2025_05_27-18_37_54")
print(PATH)
```

```python
def show_content(PATH, part):
    folder = os.path.join(PATH, part)
    for filename in sorted(os.listdir(folder)):
        content = torch.load(os.path.join(folder, filename), weights_only=True)
        print(f"{part} {filename}: \n{content.shape}")
        print(content)
        print("\n")


show_content(PATH, "Gello leader")
print("-"*100)

show_content(PATH, "Panda 10 follower")
print("-"*100)

time = os.path.join(PATH, "timestamps.pt")
timestamps = torch.load(time, weights_only=True)
print("timestamps.pt\n")
print(timestamps.shape)
print(timestamps)
# show the difference between adjacent frames
diffs = timestamps[1:] - timestamps[:-1]
print(diffs[:20])
```
